```python
f = open('sample_korea.txt')
texts=f.readlines()
```

```python
import MeCab
tokenizer=MeCab.Tagger('-d /usr/lib/mecab/dic/mecab-ko-dic/').parse
```

```python
vocab_set = set()
for i in range(len(texts)):
    text = texts[i]
    text_split = tokenizer(text).split('\n')
    
    for k in range(len(text_split)):
        token = text_split[k].split('\t')[0]
        if token != 'EOS':            
            vocab_set.add(token)
        else:
            print('EOS')
```

```python
vocab_set = list(vocab_set)
```

```python
f_out = open('vocab_korea.txt', 'w')
f_out.write('[PAD]\n')
f_out.write('[UNK]\n')
f_out.write('[CLS]\n')
f_out.write('[SEP]\n')
f_out.write('[MASK]')

for k in range(len(vocab_set)):
    f_out.write(vocab_set[k])
    if k < len(vocab_set)-1:
        f_out.write('\n')    
f_out.close()
```

```python
joo_tokenizer=MeCab.Tagger('-d /usr/lib/mecab/dic/mecab-ko-dic/').parse 
text_split = joo_tokenizer(text).split('\n')
```

```python
split_tokens = []
text_split = joo_tokenizer(text).split('\n')
words = text.split(' ')
word = ''
word_position = 0
pre_exist = 0
for k in range(len(text_split)):
    token = text_split[k].split('\t')[0]
#     print(token)
    if word_position == len(words):
        break
    
    while words[word_position] == '':
        if word_position < len(words):
            word_position += 1
        
    word += token
    if pre_exist == 0:
        split_tokens.append(token)        
        if word == words[word_position]:
            pre_exist = 0
            word_position += 1
            word = ''
        else:
            pre_exist += 1
    else:
        split_tokens.append("##"+token)
        if word == words[word_position]:
            pre_exist = 0
            word_position += 1
            word = ''
        else:
            pre_exist += 1

print(split_tokens)
```
