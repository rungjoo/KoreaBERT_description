```python
f = open('sample_korea.txt')
texts=f.readlines()
```

```python
import MeCab
tokenizer=MeCab.Tagger('-d /usr/lib/mecab/dic/mecab-ko-dic/').parse
```

```python
vocab_set = set()
for i in range(len(texts)):
    text = texts[i]
    text_split = tokenizer(text).split('\n')
    
    for k in range(len(text_split)):
        token = text_split[k].split('\t')[0]
        if token != 'EOS':            
            vocab_set.add(token)
        else:
            print('EOS')
```

```python
vocab_set = list(vocab_set)
```

```python
f_out = open('vocab_korea.txt', 'w')
f_out.write('[PAD]\n')
f_out.write('[UNK]\n')
f_out.write('[CLS]\n')
f_out.write('[SEP]\n')
f_out.write('[MASK]')

for k in range(len(vocab_set)):
    f_out.write(vocab_set[k])
    if k < len(vocab_set)-1:
        f_out.write('\n')    
f_out.close()
```
