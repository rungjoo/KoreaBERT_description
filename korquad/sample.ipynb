```python
import json
with open('KorQuAD_v1.0_dev.json', 'r') as f:
    dev = json.load(f)
    
with open('./pretrained_mecab/predictions.json', 'r') as f:
    pred = json.load(f)    
```

```python
dev_data = dev['data']
# for i in range(len(dev_data)):
for i in range(20):    
    data_samples = dev_data[i]['paragraphs']
    for k in range(len(data_samples)):
        ques = data_samples[k]['qas']
        for m in range(len(ques)):
            ans = ques[m]['answers'][0]['text']
            id = ques[m]['id']
            pred_ans = pred[id]
            print(ans, '#####', pred_ans)
            
    
```
